```python
import tensorflow as tf
from matplotlib import pyplot as plt

# linespace generates evenly spaced values in a given interval..
'''
    Arguments:
    starting index..
    ending index..
    numbers of values to generate..
'''

x = tf.linspace(-5, 5, 201)

# tf.cast is to typasting of a tensor from one data format to other..
x = tf.cast(x, tf.float32)
```

```python
def func(x):
    y = x**2 + 2*x - 5
    return y


# T is a random tensor of 201 elements
# shape = no. of elements
T = tf.random.normal(shape=[201])

y = func(x) + T

# print(x.numpy())
# print(y.numpy())
```

```python
plt.rcParams['figure.figsize'] = [9, 6]
plt.plot(x.numpy(), y.numpy(), ".", label='Data')
plt.plot(x, func(x), label="Ground Truth")
plt.legend()
plt.show()
```
